```python
import numpy as np

cities_len = 1000
def shuffle(path:np.ndarray)->None:
    np.random.shuffle(path[1:cities_len])

def shuffle_partially(path:np.ndarray)->None:
    part = np.random.choice(cities_len, 2)+1
    if part[0] < part[1]:
        np.random.shuffle(path[part[0]:part[1]])
    else:
        np.random.shuffle(path[1:part[0]])
        np.random.shuffle(path[part[1]:cities_len])

#path = np.append(np.arange(cities_len), [0])
#path[1:cities_len]
#np.random.shuffle(path[1:cities_len])
shuffle_partially(path)
path
```
