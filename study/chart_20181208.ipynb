```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.random.normal(2, 0.5, 1000)
plt.hist(x, bins=50);
```

```python
x = 30* np.random.rand(30)
y = 30* np.random.rand(30)
area = np.pi + (10*np.random.rand(30))**2
plt.scatter(x,y, s=area, marker='^', c='r');
```

```python
labels = ['baseline', 'mymodel']
data = [3.75, 4.75]
w = 0.5
xlocations = np.array(range(len(data))) + w
plt.bar(xlocations, data, width=0.5, color='b');
plt.xticks(xlocations, labels);
```

```python
df = pd.DataFrame(data={'x':x, 'y':y, 'area':area})
df.plot();
```

```python
color=['0.2', '0.5']
df.plot(kind='bar');
```

```python
df.plot(kind='box');
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16,4))
fig.suptitle('Three Graphs', size=20, y=1)
df.plot(kind='line', color=color, ax=ax1, title='Line Plot')
df.plot(kind='scatter', x='x', y='y', ax=ax2, title='Scatter Plot')
df.plot(kind='bar', color=color, ax=ax3, title='Bar Plot');
```
