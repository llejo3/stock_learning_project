```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import math
%matplotlib inline

kor_font_path = 'C:\\WINDOWS\\Fonts\\H2GTRM.TTF'
font_name = fm.FontProperties(fname=kor_font_path, size=50).get_name()
plt.rc('font', family=font_name)
```

## Box plot을 그린다.

```python
cnt = len([data_type for data_type in corps.dtypes if data_type != 'object'])
#fig, axs = plt.subplots(math.ceil(cnt/2), 2)

corps_dtypes = corps.dtypes.tolist()
corps_columns = corps.columns
for i in range(len(corps_dtypes)):
    if corps_dtypes[i] == 'object':
        i += 1
        continue
    fig, axs = plt.subplots()  
    column = corps_columns[i]
    axs.set_title(column) 
    axs.boxplot(corps[column].values, sym='', vert=0)
    fig.set_figheight(1)    
```

## Histogram

```python
cnt = len([data_type for data_type in corps.dtypes if data_type != 'object'])
#fig, axs = plt.subplots(math.ceil(cnt/2), 2)

corps_dtypes = corps.dtypes.tolist()
corps_columns = corps.columns
num_bins = 100
for i in range(len(corps_dtypes)):
    if corps_dtypes[i] == 'object':
        i += 1
        continue
    column = corps_columns[i]
    try :
        plt.hist(corps[column], num_bins, normed=1, facecolor='blue', alpha=0.5)
        plt.title(column)
        plt.show()
    except Exception as inst:
                print(inst)    
    
```
